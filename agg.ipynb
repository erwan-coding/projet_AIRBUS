```python
def signe(x):
    if x>0:
        return 1
    else:
        return -1

def has_voice_vote(X_batch,NNmodel):
    results = NNmodel.predict(X_batch,verbose=0)
    vote = 0
    for res in results:
        confiance = res[0]-res[1]
        if confiance>0:
            vote += 1
    proportion = vote/len(X_batch)
    #print(vote)
    if proportion > 0.5:
        return 1
    else:
        return 0
```

```python
X_NN = []
for k in X_train_t:
    res = has_voice_vote(k,NN)
    X_NN.append(res)
print(X_NN)
```
