```python
import librosa
import librosa.display
```

```python
amp, sample_rate = librosa.load("recordings/record_0.wav")
```

```python
hop_length = 50
```

```python
mfcc = librosa.feature.mfcc(y=amp, sr=sample_rate, hop_length=hop_length, n_mfcc=13)
```

```python
amp.shape
```

```python
spectrum = librosa.stft(y=amp)
spectrum.shape
```

```python
import matplotlib.pyplot as plt
spectrum = spectrum.astype(float)
```

```python
fig = librosa.display.specshow(spectrum, sr=sample_rate, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.plot()
plt.show()
```

```python
X = spectrum.astype(float)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)

X=scaler.transform(X) 
```

```python
from sklearn.decomposition import PCA
# Make an instance of the Model
pca = PCA(.99)
pca.fit(X)
len(pca.components_)
```

```python
import numpy as np

y = list(pca.explained_variance_ratio_)

x = range(len(y))

ycum = np.cumsum(y)

plt.bar(x,y)

plt.plot(x,ycum,"-r")

plt.show()
```
